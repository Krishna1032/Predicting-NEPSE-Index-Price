## Data analysis and Visualization

```python
data.index = pd.to_datetime(data.index, dayfirst = True)
```

Removing the zero values in the Turnover column. They are old data.

```python
data = data[data["Turnover"] != '0']
```

```python
columns_to_convert = ['Open', 'High', 'Low', 'Close','Turnover']
data[columns_to_convert] = data[columns_to_convert].replace({',': ''}, regex=True).astype(float) # remove the unnecessary  comma ','
```

```python
data.plot.line(y='Close', use_index = True, figsize = (10, 6))
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(data['Close'], model='multiplicative',period = 90)
fig = plt.figure()
fig = decomposition.plot()
fig.set_size_inches(20,7)
plt.show()
```

### Testing for stationarity using ADF and KPSS Tests.

```python
from statsmodels.tsa.stattools import adfuller, kpss

def stationarity_test(stock_close_price):
    # Calculating rolling mean and rolling standard deviation:
    rolling_mean = stock_close_price.rolling(30).mean()
    rolling_std_dev = stock_close_price.rolling(30).std()
  
    # Plotting the statistics:
    plt.figure(figsize=(20,7))
    plt.plot(stock_close_price, color='red', label='Original Time Series')
    plt.plot(rolling_mean, color='blue', label='Rolling Mean')
    plt.plot(rolling_std_dev, color='green', label='Rolling Std Dev')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show()
    
    # ADF Test
    print("ADF Test:")
    adf_test = adfuller(stock_close_price, autolag='AIC')
    print('Null Hypothesis: Not Stationary')
    print('ADF Statistic: %f' % adf_test[0])
    print('p-value: %f' % adf_test[1])
    print('Critical Values:')
    for key, value in adf_test[4].items():
        print('\t%s: %.3f' % (key, value))

    # KPSS Test
    print("KPSS Test:")
    kpss_test = kpss(stock_close_price, regression='c', nlags='auto')
    print('Null Hypothesis: Stationary')
    print('KPSS Statistic: %f' % kpss_test[0])
    print('p-value: %f' % kpss_test[1])
    print('Critical Values:')
    for key, value in kpss_test[3].items():
        print('\t%s: %.3f' % (key, value))

# Example usage:
stationarity_test(data['Close'])
```

Conclusion: This shows that the data that we have is highly non_stationary series data. The decomposition resudial component might be stationary.

```python
residual = decomposition.resid.dropna()
stationarity_test(residual)
```

## Building a model

```python
from sklearn.ensemble import RandomForestClassifier
```

Creating a target value

```python
data['Tomorrow'] = data["Close"].shift(-1) 
data["Target"] = (data["Tomorrow"] > data["Close"]).astype(int)
```

```python
data.head()
```

```python
model = RandomForestClassifier(n_estimators = 100, min_samples_split = 100, random_state = 47)
```

```python
train = data.iloc[:-100]
test = data.iloc[-100:]
```

```python
predictors = ['Open', 'High', 'Low', 'Close', 'Turnover']
```

```python
model.fit(train[predictors], train["Target"])
```

### Performance Measure

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score
```

```python
preds = model.predict(test[predictors])
```

```python
preds= pd.Series(preds, index = test.index)
preds
```

```python
accuracy_score(test["Target"], preds)
```

```python
precision_score(test["Target"], preds)
```

```python
recall_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis = 1)
combined
```
